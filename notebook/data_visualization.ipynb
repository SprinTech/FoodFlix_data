```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/home/apprenant/FoodFlix_data/Data/02_processed/nutriscore_product_cleaned.csv")
pd.set_option('max_columns', None)
```

## 1. Boxplot visualization

```python
nutriments_gr1 = ['energy_100g', 'fat_100g', 'carbohydrates_100g', 'sugars_100g']
nutrition_grade_order = ['a', 'b', 'c', 'd', 'e']
for nutriment in nutriments_gr1 :
    plt.figure(figsize = (7,7))
    sns.boxplot(x="nutrition_grade", y=nutriment, data=df, order=nutrition_grade_order)
```

```python
nutriments_gr2 = ['fiber_100g', 'proteins_100g', 'salt_100g', 'fruits-vegetables-nuts_100g']
nutrition_grade_order = ['a', 'b', 'c', 'd', 'e']
for nutriment in nutriments_gr2 :
    plt.figure(figsize = (7,7))
    sns.boxplot(x="nutrition_grade", y=nutriment, data=df, order=nutrition_grade_order)
    plt.ylim([-1,30])
```

Grâce aux informations fournies par les boxplot, on remarque que les données des produits avec un nutriscore de 'A' sont incorrectes car l'energy_100g de ces produits est bien supérieur à la moyenne des produits notés 'A' ou 'B'
